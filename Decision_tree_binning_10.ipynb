```python
import pandas as pd
from path import Path
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
file_path = Path("./Resources/bin_10_clean.csv")
df = pd.read_csv(file_path)
df.head()
```

```python
# Drop unnecessary columns 
df = df.drop(['Unnamed: 0', 'Age'], axis=1)
df
```

```python
df.dtypes
```

```python
# Binary encoding
encoded = pd.get_dummies(df, columns=["Geography"])
encoded.head()
```

```python
df.nunique()
```

```python
# Define features set
X = encoded.copy()
X = X.drop("Exited", axis=1)
X.head()
```

```python
# Define target vector
y = encoded["Exited"].values.reshape(-1, 1)
y[:5]
```

```python
# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
```

```python
# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)
```

```python
# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Creating the decision tree classifier instance
model = tree.DecisionTreeClassifier()
```

```python
# Fitting the model
model = model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data
predictions = model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Resample the training data with SMOTEENN
from imblearn.combine import SMOTEENN

sample = SMOTEENN(random_state=42)
X_resample, y_resample = sample.fit_resample(X_train_scaled, y_train)

print(y_resample)
```

```python
# Creating the decision tree classifier instance
model = tree.DecisionTreeClassifier()
```

```python
# Fitting the model
model = model.fit(X_resample, y_resample)
```

```python
# Making predictions using the testing data
predictions = model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
