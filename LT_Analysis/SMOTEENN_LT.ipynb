```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read Churn Dataset and get summary of basic statistics
df=pd.read_csv('./Resources/Churn_Modelling_2.csv')
df.describe()
```

### Before Remove Outliers

```python
#Remove columns not needed for analysis
df=df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
df
```

```python
#Encoded Gender
df['Gender']=df['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
df
```

### Smoteenn - Before Credit Score Outliers removal

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
```

```python
from pathlib import Path
from collections import Counter

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced
```

```python
#Create Features
X = pd.get_dummies(df, columns=['Geography'])
X = X.drop(["Exited"], axis=1)

# Create our target
y = df["Exited"]
```

```python
# Standardize the data with StandardScaler().
X_s= StandardScaler().fit_transform(X)
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_s, y, random_state=42)
Counter(y_train)
```

```python
# Resample the training data with SMOTEENN
# Warning: This is a large dataset, and this step may take some time to complete
from imblearn.combine import SMOTEENN

comb_sample = SMOTEENN(random_state=42)
X_resample, y_resample = comb_sample.fit_resample(X_train, y_train)

Counter(y_resample)
```

```python
# Train the Logistic Regression model using the resampled data
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resample, y_resample)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score

y_pred = model.predict(X_test)
smoteenn_accurate_score = balanced_accuracy_score(y_test, y_pred)
smoteenn_accurate_score
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm, index=["Actual Exited", "Actual Active"], columns=["Predicted Exited", "Predicted Active"])

cm_df
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(f'smoteenn classification report')
print(classification_report_imbalanced(y_test, y_pred))
```

### Remove Outliers

```python
#Remove Credit Score Outliers in a new DataFrame
import numpy as np
Q3 = np.quantile(df['CreditScore'], 0.75)
Q1 = np.quantile(df['CreditScore'], 0.25)
IQR = Q3 - Q1
    
lower_range = Q1 - 1.5 * IQR
upper_range = Q3 + 1.5 * IQR
outlier_free_list = [x for x in df['CreditScore'] if (
    (x > lower_range) & (x < upper_range))]
filt_dt = df.loc[df['CreditScore'].isin(outlier_free_list)]
```

```python
#Check new DataFrame after removing Credit Score outliers
filt_dt
```

### Smoteenn Remove Credit Score Outliers

```python
#Create Features
X = pd.get_dummies(filt_dt, columns=['Geography'])
X = X.drop(["Exited"], axis=1)

# Create our target
y = filt_dt["Exited"]
```

```python
# Standardize the data with StandardScaler().
X_s= StandardScaler().fit_transform(X)
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_s, y, random_state=42)
Counter(y_train)
```

```python
# Resample the training data with SMOTEENN
# Warning: This is a large dataset, and this step may take some time to complete
from imblearn.combine import SMOTEENN

comb_sample = SMOTEENN(random_state=42)
X_resample, y_resample = comb_sample.fit_resample(X_train, y_train)

Counter(y_resample)
```

```python
# Train the Logistic Regression model using the resampled data
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resample, y_resample)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score

y_pred = model.predict(X_test)
smoteenn_accurate_score = balanced_accuracy_score(y_test, y_pred)
smoteenn_accurate_score
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm, index=["Actual Exited", "Actual Active"], columns=["Predicted Exited", "Predicted Active"])

cm_df
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(f'smoteenn classification report')
print(classification_report_imbalanced(y_test, y_pred))
```
